### Data Preparation and Merging

```python
import pandas as pd
import os
```

```python
os.makedirs("prepared_dataset", exist_ok=True)
```

## Hourly Steps dataset

```python
Steps_hr_1 = pd.read_csv("../archive/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/hourlySteps_merged.csv")
Steps_hr_2 = pd.read_csv("../archive/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/hourlySteps_merged.csv")
```

```python
Steps_hr = pd.concat([Steps_hr_1, Steps_hr_2], ignore_index=True) 
## Concatenate the datasets
```

```python
print(Steps_hr.shape)

Steps_hr.head()
```

```python
print(Steps_hr.columns)
```

```python
Steps_hr.info()
```

```python
Steps_hr.rename(columns={
    'ActivityHour': 'datetime',

}, inplace=True)
```

```python

Steps_hr['datetime'] = pd.to_datetime(Steps_hr['datetime'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
Steps_hr.head()
```

## Hourly Sleep Dataset

```python
sleep_1= pd.read_csv("../archive/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/minuteSleep_merged.csv")
sleep_2 = pd.read_csv("../archive/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/minuteSleep_merged.csv")
```

```python
sleep = pd.concat([sleep_1, sleep_2], ignore_index=True)
```

```python
print(sleep.shape)

sleep.head()
```

```python
sleep.info()
```

```python
sleep.rename(columns={
    'date': 'datetime',

}, inplace=True)
```

```python


sleep['datetime'] = pd.to_datetime(sleep['datetime'])


sleep['datetime'] = sleep['datetime'].dt.floor('H')

sleep_hourly = sleep.groupby(['Id', 'datetime', 'value']).size().unstack(fill_value=0).reset_index()


sleep_hourly.columns.name = None
sleep_hourly.rename(columns={1: 'asleep_min', 2: 'restless_min', 3: 'awake_min'}, inplace=True)


sleep_hourly['total_sleep_min'] = sleep_hourly.get('asleep_min', 0) + sleep_hourly.get('restless_min', 0)


```

```python
sleep.head()
```

## Hourly Calories

```python
Calories_hr_1 = pd.read_csv("../archive/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/hourlyCalories_merged.csv")
Calories_hr_2 = pd.read_csv("../archive/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/hourlyCalories_merged.csv")
```

```python
Calories_hr = pd.concat([Calories_hr_1, Calories_hr_2], ignore_index=True)
```

```python
Calories_hr.info()
```

```python
print(Calories_hr.shape)

Calories_hr.head()
```

```python
Calories_hr.rename(columns={
    'ActivityHour': 'datetime',

}, inplace=True)
```

```python
Calories_hr['datetime'] = pd.to_datetime(Calories_hr['datetime'])
```

```python
Calories_hr.head()
```

## Hourly Heart Rate

```python

hr_1 = pd.read_csv("../archive/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/heartrate_seconds_merged.csv")
hr_2 = pd.read_csv("../archive/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/heartrate_seconds_merged.csv")
```

```python
hr = pd.concat([hr_1, hr_2], ignore_index=True)
```

```python
hr.info()
```

```python
print(hr.shape)

hr.head()
```

```python
hr.rename(columns={
    'Time': 'datetime',

}, inplace=True)
```

```python

hr['datetime'] = pd.to_datetime(hr['datetime'])


hr['datetime'] = hr['datetime'].dt.floor('H')

hr_hourly = hr.groupby(['Id', 'datetime'])['Value'].mean().reset_index()


hr_hourly.rename(columns={'Value': 'hourly_heart_rate'}, inplace=True)
```

```python

merged = pd.merge(Steps_hr, sleep_hourly, on=['Id', 'datetime'], how='outer')
merged = pd.merge(merged, hr_hourly, on=['Id', 'datetime'], how='outer')
merged = pd.merge(merged, Calories_hr, on=['Id', 'datetime'], how='outer')
```

```python
merged.info()
```

```python
merged.head()
```

```python
merged.shape
```

```python
merged.isnull().sum()
```

```python
merged.fillna({'StepTotal': 0, 'total_sleep_min': 0, 'Calories':0 }, inplace=True)
```

```python
merged.isnull().sum()
```

```python
merged = merged.drop(columns=['asleep_min', 'restless_min', 'awake_min' ])
```

```python
merged.isnull().sum()
```

```python
df_clean = merged.dropna()
```
